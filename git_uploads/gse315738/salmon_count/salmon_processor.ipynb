```python
import pandas as pd
import glob

# 1. Identify and merge files
files = [f for f in glob.glob("qc_log*.csv") if "merged" not in f]
merged = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)

# 2. Identify IDENTICAL duplicates (Every single column matches)
identical_mask = merged.duplicated(keep=False) # Flags all rows that have a twin
identical_count = merged.duplicated().sum()

# 3. Identify SRR duplicates that are NOT identical (Same SRR, different data)
# This checks if the SRR is repeated but the rest of the row might be different
srr_repeats = merged.duplicated(subset=['srr'], keep=False)
non_identical_repeats = merged[srr_repeats & ~identical_mask]

# 4. Remove the identical ones (Keep only the first occurrence)
clean_df = merged.drop_duplicates()

# 5. Sort by SRR
clean_df = clean_df.sort_values(by='srr')

# 6. Save
clean_df.to_csv("qc_log_merged.csv", index=False)

# --- Reporting ---
print(f"Initial rows: {len(merged)}")
print(f"Identical duplicate rows removed: {identical_count}")

if not non_identical_repeats.empty:
    print(f"\n⚠️ WARNING: Found {len(non_identical_repeats)} rows with the same SRR but DIFFERENT data:")
    print(non_identical_repeats[['srr', 'timestamp', 'salmon_mapping_rate_pct']])
else:
    print("\n✅ All duplicates found were 100% identical and have been cleaned.")

print(f"\nFinal unique file saved: {clean_df.shape}")
```
