```python
import pandas as pd
import glob

files = [f for f in glob.glob("qc_log*.csv") if "merged" not in f]

for f in files:
    df = pd.read_csv(f)
    print(f"\n{'='*60}")
    print(f"FILE: {f}  ({len(df)} rows)")
    print('='*60)
    print(df.to_string())

merged = pd.concat([pd.read_csv(f) for f in files], ignore_index=True)

print(f"\n{'='*60}")
print(f"MERGED OUTPUT  ({len(merged)} rows from {len(files)} files)")
print('='*60)
print(merged.to_string())
```
