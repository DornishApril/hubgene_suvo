```python
import os, json, subprocess, shutil

# ── CONFIG ──
KAGGLE_USERNAME = "your_kaggle_username"
KAGGLE_API_KEY  = "your_api_key_here"

# ── 1. Setup API key ──
os.makedirs("/root/.kaggle", exist_ok=True)
with open("/root/.kaggle/kaggle.json", "w") as f:
    json.dump({"username": KAGGLE_USERNAME, "key": KAGGLE_API_KEY}, f)
os.chmod("/root/.kaggle/kaggle.json", 0o600)
subprocess.run(["pip", "install", "-q", "kaggle"], check=True)
print("API key ready.")

# ── 2. Create a small test file in /kaggle/temp ──
os.makedirs("/kaggle/temp/test_upload", exist_ok=True)
with open("/kaggle/temp/test_upload/hello.txt", "w") as f:
    f.write("This file was created in /kaggle/temp and uploaded directly.\n")
print("Test file created in /kaggle/temp.")

# ── 3. Write dataset metadata ──
meta = {
    "title": "Test Upload From Temp",
    "id":    f"{KAGGLE_USERNAME}/test-upload-from-temp",
    "licenses": [{"name": "other"}]
}
with open("/kaggle/temp/test_upload/dataset-metadata.json", "w") as f:
    json.dump(meta, f)

# ── 4. Upload directly from /kaggle/temp ──
print("Uploading...")
r = subprocess.run([
    "kaggle", "datasets", "create",
    "-p", "/kaggle/temp/test_upload",
    "--dir-mode", "zip"
], capture_output=True, text=True)

print("STDOUT:", r.stdout)
print("STDERR:", r.stderr)

if r.returncode == 0:
    print(f"\n✓ Success! Check: https://www.kaggle.com/datasets/{KAGGLE_USERNAME}/test-upload-from-temp")
else:
    raise RuntimeError("Upload failed — check STDERR above.")

# ── 5. Cleanup ──
shutil.rmtree("/kaggle/temp/test_upload")
print("Cleaned up.")
```
